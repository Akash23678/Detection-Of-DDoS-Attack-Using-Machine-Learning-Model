```python
# firstly upload raw dataset that collected from wireshark
# Uplaod required library

import numpy as np
import pandas as pd
```

```python
# Load the Dataset

df=pd.read_csv("/content/tcp_syn.csv")
df
```

```python
# Drop the unnamed column if present
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df
```

```python
# There is a complex Info column so we need to extract some important features like "SYN", "ACK"
# Extract 'SYN' or 'ACK' from 'Info' column

df['Info'] = df['Info'].apply(lambda x: 'SYN' if '[SYN]' in x else ('ACK' if '[SYN, ACK]' in x else 'Other'))  # if col contains string [SYN] then update it to SYN and if contains [SYN, ACK] then update to ACK and otherwise update to Other
df
```

Add new column for labelling

```python
df.head()
```

```python
# Create Class Label

df["Result"] = ""          # create new "Result" class label

for i in range(len(df)-1):          
    
    # Check if current row is SYN and next row is not ACK
    if df.iloc[i]["Info"] == "SYN" and df.iloc[i+1]["Info"] != "ACK":
        df.at[i, "Result"] = "malicious"        # then we conclude that it is malicious so set/update result=malicious
    else:
        df.at[i, "Result"] = "normal"
        
# Label the last row as normal
df.at[len(df)-1, "Result"] = "normal"



df
```

```python
df.loc[df["Protocol"] == "ARP" , "Source"] = "192.168.66.128"
df.loc[df["Protocol"] == "ARP", "Destination"] = "192.168.66.129"
df
```

```python
# Drop unnecessary col from dataset
df=df.drop(['Time','Destination Port','Protocol','Length'], axis=1)
df
```

```python
#to update the column "Source"

df.loc[(df["Info"]  == "SYN") | (df["Info"] == "Other"), "Source"] = 100  # Here 100 take it as user ip 
df.loc[df["Info"] == "ACK", "Source"] = 200   # Here 200 take it as server ip


#to update the column "Destination"
df.loc[(df["Info"]  == "SYN") | (df["Info"] == "Other"), "Destination"] = 200  #SYN->server(200)
df.loc[df["Info"] == "ACK", "Destination"] = 100   #ACK->user(100)
# df=df.drop(['Info'], axis=1)
df
```

```python
df.loc[df["Info"].str.contains("ACK|Other"), "Info"] = 0            
df.loc[df["Info"] == "SYN", "Info"] = 1
df
```
